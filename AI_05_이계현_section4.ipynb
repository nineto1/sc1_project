```python
from os import listdir
from PIL import Image
from numpy import asarray
from matplotlib import pyplot
from mtcnn.mtcnn import MTCNN
```

```python
def extract_face(filename, required_size=(160, 160)) : 
  images = []
  image = Image.open(filename)
  image = image.convert('RGB')
  pixels = asarray(image)
  detector = MTCNN()
  results = detector.detect_faces(pixels)
  for result in results :
      if result['confidence'] < 0.9 : continue
      x1, y1, width, height = result['box']
      x1, y1 = abs(x1), abs(y1)
      x2, y2 = x1 + width, y1 + height
      face = pixels[y1:y2, x1:x2]
      image = Image.fromarray(face)
      image = image.resize(required_size)      
      images.append(image)  
  return images
```

```python
folder = "/content/drive/MyDrive/download"
save_folder = "/content/drive/MyDrive/face"

for dir in listdir(folder) :
  i = 0
  print(dir)  
  sub_dir = folder + "/" + dir
  
  for filename in listdir(sub_dir) :
    path = sub_dir + "/" + filename
    faces = extract_face(path)

    if len(faces) == 0 : continue

    for face in faces :
      save_path = save_folder + "/" + dir + "/" + str(i) + ".jpg" 
      face.save(save_path)
      i += 1
    
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf

image_gen = ImageDataGenerator(rescale=1./255, validation_split = 0.2)

train = image_gen.flow_from_directory(
        '/content/drive/MyDrive/face',
        target_size=(160, 160),
        batch_size=10000,        
        subset = 'training',
        class_mode='binary')

test = image_gen.flow_from_directory(
        '/content/drive/MyDrive/face',
        target_size=(160, 160),
        batch_size=10000,
        subset='validation',
        class_mode='binary')
```

```python
x_train = train[0][0]
y_train = train[0][1]
x_test = test[0][0]
y_test = test[0][1]
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout, Dense, Conv2D, MaxPooling2D, Flatten

model = Sequential() 
model.add(Conv2D(64, (3,3), activation='relu', input_shape=(160,160,3)))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(32, (3,3), activation='relu'))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = 'accuracy')
```

```python
model.fit(x_train, y_train, batch_size=32, epochs=200, validation_data=(x_test, y_test))
```

```python
def model_builder(con_chann=[32,64,64], dense_node=[64,32], dropout=0.2):
  model = Sequential() 
  model.add(Conv2D(con_chann[0], (3,3), activation='relu', input_shape=(160,160,3)))
  model.add(MaxPooling2D((2,2)))
  model.add(Conv2D(con_chann[1], (3,3), activation='relu'))
  model.add(MaxPooling2D((2,2)))
  model.add(Conv2D(con_chann[2], (3,3), activation='relu'))
  model.add(Flatten())
  model.add(Dense(dense_node[0], activation='relu'))
  model.add(Dropout(dropout))
  model.add(Dense(dense_node[1], activation='relu'))
  model.add(Dropout(dropout))
  model.add(Dense(1, activation='sigmoid'))
  model.summary()
  model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = 'accuracy')
  return model
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

hypermodel = KerasClassifier(build_fn=model_builder, verbose=0)

# GridSearch
epochs = [100]
batch_size = [32, 128]
con_chann = [[64, 64, 64], [16, 16, 16]]
dense_node = [[32, 32], [16, 16]]
dropout = [0.2, 0.5, 0.7]
param_grid = dict(batch_size=batch_size, epochs=epochs, con_chann=con_chann, dense_node=dense_node, dropout=dropout)


# GridSearch CV를 만들기
grid = GridSearchCV(estimator=hypermodel, param_grid=param_grid, cv=3, verbose=2, n_jobs=-1)
grid_result = grid.fit(x_train, y_train)
```

```python
print(f"Best: {grid_result.best_score_} using {grid_result.best_params_}")
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print(f"Means: {mean}, Stdev: {stdev} with: {param}")
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

best_model = grid_result.best_estimator_
y_pred = best_model.predict(x_test)
print(f'deep learning test accuracy score : {accuracy_score(y_test, y_pred)}')
```

# 3. 한계점과 개선방안.
- 데이터 수집의 한계 : 이번 프로젝트의 부수적인 목적은 AI 를 이용하여 빠르게 필요한 데이터를 확보하는 것이 었으나, 단순히 구글 검색만으로 label 에 맞는 정확한 이미지 수집의 한계를 느낌.
- 단순 모델 훈련 시 과적합되는 현상이 보였고, 하이퍼 파라미터 튜닝을 시도 하였으나, 기대만큼의 개선효과가 없는 것으로 보임.
- 정확한 모델훈련을 위해 우선 label 에 맞는 정확한 훈련데이터 확보를 선행할 필요가 있으며 이후 모델에 성능개선이 필요할 것으로 보임.
- 충분한 컴퓨팅 resource 가 있는 경우, 얼굴 이미지 추출을 전처리 과정이 아닌 모델의 일부로 편입하고 일반 이미지를 input 으로 하여 전체 모델을 통일 할 수 있음.
